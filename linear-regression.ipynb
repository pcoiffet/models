```python
df_train = pd.read_csv("/kaggle/input/random-linear-regression/train.csv")
df_test = pd.read_csv("/kaggle/input/random-linear-regression/test.csv")

df_train = df_train.dropna(subset=['y'])

df_train.shape
```

```python
x = df_train[['x']]
y = df_train['y']
```

```python
import matplotlib.pyplot as plt 

plt.scatter(x,y,s=10)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Apperçu des données")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
model = LinearRegression()
model
```

```python
model.fit(x,y)
```

```python
predictions = model.predict(df_test[["x"]])
```

```python
mse = mean_squared_error(df_test['y'], predictions)
```

```python
mse
```

```python
plt.scatter(x,y,s=10,label="données réelles")
plt.plot(df_test[['x']],predictions,color='red',label = "droite de régression")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Apperçu des données")

plt.show()
```

```python
r2 = model.score(df_test[['x']], df_test['y'])
print("R2:", r2)
```
